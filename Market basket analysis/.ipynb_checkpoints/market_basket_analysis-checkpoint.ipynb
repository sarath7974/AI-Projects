# Importing Packages

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

# Read the data

```python
data = pd.read_excel("online_retail_data.xlsx",engine = "openpyxl")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

# Data Preprocessing

```python
data=data.dropna()                                    #drop null values
data["Description"]=data["Description"].str.strip()   #remove spaces from description
data.drop(data[data["Quantity"]<=0].index)            #remove negative values from quantity
data['Invoice']=data["Invoice"].astype("str")         #convert invoice to string type
data=data[~data["Invoice"].str.contains("C")]         #uses bitwise negation operation to remove all the credit transactions
```

```python
data.shape
```

```python
data["Country"].value_counts()
```

```python
count= data['Description'].unique()
len(count)
```

```python
# Most bought top 10 items
import matplotlib.pyplot as plt
most_bought = data.groupby("Description")["Quantity"].sum().sort_values(ascending=False)[:10]
most_bought = most_bought.to_frame()   #convert series to dataframe
most_bought.reset_index(inplace = True)
most_bought.plot(kind='bar',x='Description',y='Quantity',figsize=(16,8),title="Top 10 Most Bought Items")
```

# Create Basket Data

```python
mybasket = (data[data['Country'] =="United Kingdom"]
          .groupby(['Invoice', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Invoice'))

mybasket.head()
```

# Encode the Data

### Converting positive values to 1 and negative values to 0

```python
#converting +ve values to 1 and -ve values to 0
def encoder(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_data = mybasket.applymap(encoder)                        #Apply encoding to dataframe
basket_data.drop("POSTAGE",inplace=True,axis=1)                   
basket_data = basket_data[(basket_data > 0).sum(axis=1) >= 2]   #filter out the transactions that have bought more than 1 items only
basket_data
```

# Training the model

### Generate Frequent Itemset

```python
def frequent(data_items):
    frequent = apriori(data_items, min_support=0.03, use_colnames=True)
    frequent.sort_values("support", axis = 0, ascending = False,inplace = True)
    frequent.reset_index(drop=True,inplace = True)
    return frequent
```

```python
frequent_itemsets = frequent(basket_data)
```

```python
#Visualize Frequently bought items
frequent_plot = frequent_itemsets.head(10)
frequent_plot.plot(kind="bar",x="itemsets",y='support',figsize=(16,8),title="Top 10 Frequently Bought items")
```

```python
frequent_itemsets.head()
```

### Find association between fequently bought items

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```
